### 1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# list of 5 integers
integer_list = [125, 43, 102, 237, 23]
print(integer_list)
```

```python
# append operation (adds to end of list)
integer_list.append(200)
print(integer_list)
```

```python
# extend operation - appends each element from argument list to end of integer_list
integer_list.extend([10,200])
print(integer_list)
```

```python
# insert operation - first argument is position in list, second is value to insert 
# this will insert in index 2 (3rd item in list)
integer_list.insert(2, 10)
```

```python
print(integer_list)
```

```python
# remove operation - removes the very first occurance of value that is passed as argument
integer_list.remove(10)
print(integer_list)
integer_list.remove(102)
print(integer_list)
```

```python
# remove last element at end of list
integer_list.pop()
print(integer_list)
```

```python
# clear - removes all elements from list
integer_list.clear()
print(integer_list)
```

```python
# index - returns the first index position of that element from that list
integer_list = [200, 12, 13, 12, 10, 90, 67, 10, 10]
print(integer_list.index(10)) # should be index 4
print(integer_list.index(12)) # should be index 1
```

```python
# count - returns the number of elements with specified value (passed as argument)
print(integer_list.count(10)) # should print 3
print(integer_list.count(12)) # should print 2
```

```python
# reverse method is used to reverse the order of the list
integer_list.reverse()
print(integer_list) # should print [10, 10, 67, 90, 10, 12, 13, 12, 200]
```

### 5. Write a script that checks if a person is eligible to vote (age ≥ 18). Take age as a variable and print the appropriate message.

```python
def is_able_to_vote(age: int) -> bool:
    if age < 0:
        raise ValueError("Negative age is not allowed. Please try again")
    return age >= 18
```

```python
print(is_able_to_vote(10)) # less than 18, false
print(is_able_to_vote(18)) # equal to 18, true
print(is_able_to_vote(50)) # greater than 18, true
```

### 6. Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
def print_grades(marks: int) -> str: 
    # order of if statements matter!
    # elif - else if i.e. "otherwise" => only executes this block if previous if statement is false
    # starting with Type Check and Value Error as marks are most likely between 0 - 100 (integer)
    # then goes on to check if >= 90, else check between 75-89, else check between 50-74, else check below 50
    if type(marks) is not int:
        raise TypeError("Marks are an integer grade from 0 to 100, please pass in an argument that is an integer")
    elif marks > 100:
        raise ValueError("Marks cannot be above 100. Marks are an integer grade from 0 to 100")
    elif 90 <= marks <= 100:
        print('A')
    elif marks >= 75:
        print('B')
    elif marks >= 50:
        print('C')
    else:
        print('Fail')
```

```python
# Test
print_grades("Abhinav Thota") # should raise TypeError
```

```python
print_grades(120) # Should raise ValueError
```

```python
# test

print_grades(30) # Fail
print_grades(49) # Fail
print_grades(55) # C
print_grades(75) # B
print_grades(90) # A
print_grades(99) # A
```

#### 7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
# wrote simple one line fns
def is_even(num : int) -> bool:
    '''checks if number is even'''
    # modulo returning 0 would mean no remainder when divided by two, hence number is even
    # else it would return 1, and that would be false (not even)
    return num % 2 == 0
def is_positive(num : int) -> bool:
    '''checks if number is positive'''
    return num > 0
def is_zero(num: int) -> bool:
    '''checks if number is zero'''
    return num == 0
def is_negative(num: int) -> bool:
    '''checks if number is negative'''
    return num < 0

def main(num : int):
    if is_positive(num):
        print (f"{num} is positive")
    if is_negative(num):
        print(f"{num} is negative")
    if is_zero(num):
        print(f"Number is zero")
    if is_even(num):
        print(f"{num} is even")


```

```python
# test
main(0)
main(1)
main(120)
main(-2)
```
